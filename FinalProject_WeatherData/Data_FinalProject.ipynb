# Geocode stations

```python
#Create df with station number and location(street)
StaLoc_df=StaNum_df[['Start station number', 'Start station']]
StaLoc_df.head()
```

```python
pd.set_option("max_rows", None)
pd.set_option("max_colwidth",None)
StaLoc_df
```

```python
#Prepare address location for API
StaLoc_df["Address"]= StaLoc_df["Start station"].str.replace("&", "and", case = False)
```

```python
#Display df
StaLoc_df.head()
```

```python
#Prepare address location for API
StaLoc_df["Add"]= StaLoc_df["Address"].str.replace("/", "and", case = False)
```

```python
#Display df
StaLoc_df
```

```python
#Add city
StaLoc_df['Location'] = StaLoc_df["Add"].astype(str) + ', Washigton, DC'
StaLoc_df
```

```python
#Delete no used columns
del StaLoc_df['Start station']
del StaLoc_df['Address']
del StaLoc_df['Add']
```

```python
StaLoc_df
```

```python
#create list of address
StaLoc_list = StaLoc_df["Location"].tolist()
StaLoc_list
```

```python
#Do API to google maps and store lat and lng for each station
lat = []
lng = []

for add in StaLoc_list:
    # Target add
    target= add

    # Build the endpoint URL
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target, gkey)
    geo_data = requests.get(target_url).json()
    latitud = geo_data["results"][0]["geometry"]["location"]["lat"]
    longitud = geo_data["results"][0]["geometry"]["location"]["lng"]
    lat.append(latitud)
    lng.append(longitud)
    print('''
    City: {0}
    Latitude: {1}
    Longitude: {2}
    '''.format(target, latitud, longitud))
```

```python
#Add latitud column to df
StaLoc_df["Latitud"]=lat
```

```python
#Add longitud column to df
StaLoc_df["Longitud"]=lng
```

```python
#Display final df witn station number addres and location
StaLoc_df
```

```python
#Export to csv
StaLoc_df.to_csv("CapitalBikeShare_Stations_2011.csv")
```
